```python
import pandas as pd
import numpy as np

from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier

from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score,
    confusion_matrix,
    classification_report
)
```

```python
X_train = pd.read_csv("../data/processed/X_train.csv")
X_test = pd.read_csv("../data/processed/X_test.csv")

y_train = pd.read_csv("../data/processed/y_train.csv").squeeze("columns")
y_test = pd.read_csv("../data/processed/y_test.csv").squeeze("columns")
```

```python
def evaluate_model(model_name,model, X_test, y_test):
    
    y_pred = model.predict(X_test)
    y_prob = model.predict_proba(X_test)[:, 1]

    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    roc_auc = roc_auc_score(y_test, y_prob)

    return {
        "Model Name : ":model_name,
        "Accuracy": accuracy,
        "Precision": precision,
        "Recall": recall,
        "F1 Score": f1,
        "ROC AUC": roc_auc
    }
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
```

```python
knn_result = evaluate_model("KNN",knn,X_test,y_test)
knn_result
```

```python
all_result = []
```

```python
all_result.append(knn_result)
```

```python
all_result
```

```python
comparison_df = pd.DataFrame(all_result)
comparison_df = comparison_df.round(4)
comparison_df
```

## Model Comparison Summary

Five machine learning models were evaluated using Accuracy, Precision, Recall, F1-score, and ROC-AUC.

### Key Findings

- KNN achieved the highest Accuracy (83.9%), Precision (66.8%), and F1-score (51.2%), making it the strongest overall performer.
- Balanced Logistic Regression achieved the highest Recall (70.0%), making it effective at identifying churned customers.
- Naive Bayes achieved the highest ROC-AUC (0.7843), indicating good ranking capability.
- Decision Tree provided moderate performance but is likely to benefit from hyperparameter tuning.
- Logistic Regression served as a strong baseline model but struggled to identify churned customers due to low Recall.

### Conclusion

There is no universally best model. The appropriate model depends on the business objective:

- If minimizing missed churned customers is the priority, Balanced Logistic Regression is preferred.
- If achieving the best overall balance across multiple metrics is the priority, KNN is the strongest candidate among the evaluated models.

# Hyperparameter Tuning

```python
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn_params = {
    "n_neighbors": [3,5,7,9,11,13,15]
}
```

### Creating Grid Search

```python
grid_knn = GridSearchCV(
    estimator = KNeighborsClassifier(),
    param_grid = knn_params,
    cv = 5,
    scoring = "f1",
    n_jobs = -1
)
```

```python
grid_knn.fit(X_train, y_train)
```

```python
grid_knn.best_params_
```

```python
grid_knn.best_score_
```

```python
cv_results = pd.DataFrame(grid_knn.cv_results_)

cv_results[
    [
        "param_n_neighbors",
        "mean_test_score",
        "rank_test_score"
    ]
].sort_values(
    by="rank_test_score"
).head(10)
```

## KNN Hyperparameter Tuning

GridSearchCV was used with 5-fold Cross Validation to identify the optimal value of `n_neighbors`.

### Results

- Best Hyperparameter:
  - n_neighbors = 3

- Best Cross Validation F1 Score:
  - 0.5131

### Observation

- The default value (n_neighbors = 5) was not optimal for this dataset.
- GridSearchCV selected n_neighbors = 3 based on the highest mean F1-score across the validation folds.
- Cross Validation provides a more reliable estimate of model performance than a single train-validation split.

```python
#Creating a new KNN model with the best parameters from GridSearchCV
best_knn = KNeighborsClassifier(n_neighbors=grid_knn.best_params_['n_neighbors'])
```

```python
best_knn.fit(X_train, y_train)
```

```python
best_knn_res = evaluate_model("Best KNN",best_knn,X_test, y_test)
best_knn_res
```

## KNN Hyperparameter Tuning Results

GridSearchCV selected `n_neighbors = 3` as the optimal parameter based on 5-fold Cross Validation.

However, when evaluated on the unseen test set, the tuned model did not outperform the default KNN model across most evaluation metrics.

### Comparison

- Recall improved slightly.
- Accuracy decreased.
- Precision decreased.
- F1-score decreased slightly.
- ROC-AUC decreased.

### Conclusion

Although GridSearchCV identified the best parameter on the training data, the default KNN (k=5) generalized better on the test set. Therefore, the default KNN model is retained for further comparison.
